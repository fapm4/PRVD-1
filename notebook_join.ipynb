### YRBS 2023

```python
import pyodbc
import pandas as pd

# Ruta al archivo .mdb
mdb_file = "datos/XXH2023_YRBS_Data.mdb"

# Crear la conexión con pyodbc
conn = pyodbc.connect(
    r"Driver={Microsoft Access Driver (*.mdb, *.accdb)};"
    f"DBQ={mdb_file};"
)

# Consulta SQL para leer datos de una tabla específica
query_q = "SELECT * FROM XXHq"  # set de preguntas Q


# Ejecutar la consulta y cargar los datos en un DataFrame
df = pd.read_sql(query_q, conn)


# Cerrar la conexión
conn.close()

# Mostrar las primeras filas del DataFrame
df.head()
```

```python
# Cogemos las variables que vamos a estudiar
df_y23 = df[['q1', 'q2', 'q3', 'q25', 'q26', 'q27', 'q29', 'q42', 'q80', 'q84', 'q85', 'q87']]
df_y23.columns = ['age', 'gender', 'school_grade', 'electronic_bullying', 'stop_daily_activities', 'consider_suicide', 'attempt_suicide', 'current_alcohol', 'social_use', 'depressed', 'sleep', 'academic_grades']
df_y23['Year'] = 2023
df_y23
```

```python
# Rellenar nulos con 0 para convertir a entero
# df_y = df_y.fillna(0)
# df_y = df_y.astype('int32')
```

```python
# Diccionarios de mapeo para cada columna
map_age = {
    '1': '<=12',  
    '2': '13',
    '3': '14',
    '4': '15',
    '5': '16',
    '6': '17',
    '7': '>=18'
}

map_gender = {'1': 'Female', '2': 'Male'}

map_schoolGrade = {
    '1': 9,  
    '2': 10,
    '3': 11,
    '4': 12,
    '5': 'Ungraded or other grade'
}

map_bullying = {'1': 'Yes', '2': 'No'}

map_dailyActivities = {'1': 'Yes', '2': 'No'}

map_ConsiderSuicide = {'1': 'Yes', '2': 'No'}

map_attemptSuicide = {'1': '0 times', '2': '1 time', '3': '2-3 times', '4': '4-5 times', '5': '6 or more times'}

map_alcohol = {'1': '0 days', '2': '1-2 days', '3': '3-5 days', '4': '6-9 days', '5': '10-19 days', '6': '20-29 days', '7': 'All 30 days'}

map_social = {
    '1': 'I do not use social media',
    '2': 'A few times a month',
    '3': 'About once a week',
    '4': 'A few times a week',
    '5': 'About once a day',
    '6': 'Several times a day',
    '7': 'About once an hour',
    '8': 'More than once an hour'
}

map_depressed = {'1': 'Never', '2': 'Rarely', '3': 'Sometimes', '4': 'Most of the time', '5': 'Always'}

map_sleep = {'1': '<=4h', '2': '5h', '3': '6h', '4': '7h', '5': '8h', '6': '9h', '7': '>=10h'}

map_academic = {'1': 'A', '2': 'B', '3': 'C', '4': 'D', '5': 'F', '6': 'Other', '7': 'Unsure'}

# Crear un diccionario general para mapear cada columna
mapeos = {
    'age': map_age,
    'genre': map_gender,
    'electronic_bullying': map_bullying,
    'social_use': map_social
}


# Aplicar los mapeos a las columnas del DataFrame
for columna, mapeo in mapeos.items():
    df_y23[columna] = df_y23[columna].replace(mapeo)

# Mostrar el DataFrame con los valores categóricos
df_y23.head()
```
