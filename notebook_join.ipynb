## Unión

```python
df_k.dtypes
```

```python
# Diccionarios de mapeo para cada variable

# Edad
map_age = {
    '1': 12,  
    '2': 13,
    '3': 14,
    '4': 15,
    '5': 16,
    '6': 17,
    '7': 18
}
df_y21['age_normalized'] = df_y21['age'].map(map_age)
df_y23['age_normalized'] = df_y23['age'].map(map_age)
df_k['age_normalized'] = df_k['age'].astype('int32')

# Género
map_gender = {'1': 'Female', '2': 'Male'}
df_y21['gender_normalized'] = df_y21['gender'].map(map_gender)
df_y23['gender_normalized'] = df_y23['gender'].map(map_gender)


# School grade y occupation
schoolGrade_to_occupation = {
    '1': 'School Student', #'9th grade',  
    '2': 'School Student', #'10th grade',
    '3': 'School Student', #'11th grade',
    '4': 'School Student', #'12th grade',
    '5': 'Salaried Worker' #'Ungraded or other grade'
}
df_y21['occupation_normalized']  = df_y21['school_grade'].map(schoolGrade_to_occupation)
df_y23['occupation_normalized']  = df_y23['school_grade'].map(schoolGrade_to_occupation)


# Stopped doing daily activities
map_dailyActivities = {'1': 'Yes', '2': 'No'}
df_y21['stop_daily_activities_normalized']  = df_y21['stop_daily_activities'].map(map_dailyActivities)
df_y23['stop_daily_activities_normalized']  = df_y23['stop_daily_activities'].map(map_dailyActivities)

scale_to_dailyActivities = { 1: 'No', 2: 'No', 3: 'Yes', 4: 'Yes', 5: 'Yes'}
df_k['stop_daily_activities_normalized'] = df_k['stop_daily_activities'].map(scale_to_dailyActivities)


# Social media use
frequency_to_hours_social = {
    '1': '0 hours', #'I do not use social media',
    '2': 'Less than an Hour', #'A few times a month',
    '3': 'Less than an Hour', #'About once a week',
    '4': 'Less than an Hour', #'A few times a week',
    '5': 'Between 1 and 2 hours', #'About once a day',
    '6': 'Between 2 and 3 hours', #'Several times a day',
    '7': 'Between 3 and 4 hours', #'About once an hour',
    '8': 'More than 5 hours' #'More than once an hour'
}
df_y23['social_use_normalized'] = df_y23['social_use'].map(frequency_to_hours_social)

map_social_use = {
    '1': 'Less than an Hour',
    '2': 'Between 1 and 2 hours',
    '3': 'Between 2 and 3 hours',
    '4': 'Between 3 and 4 hours',
    '5': 'Between 3 and 4 hours',
    '6': 'More than 5 hours'
}
df_y21['social_use_normalized'] = df_y21['social_use'].map(map_social_use)


# Depressed
map_depressed = {'1': 'Never', '2': 'Rarely', '3': 'Sometimes', '4': 'Most of the time', '5': 'Always'}
map_depressed_int = {1: 'Never', 2: 'Rarely', 3: 'Sometimes', 4: 'Most of the time', 5: 'Always'}
df_y21['depressed_normalized'] = df_y21['depressed'].map(map_depressed)
df_y23['depressed_normalized'] = df_y23['depressed'].map(map_depressed)
df_k['depressed_normalized'] = df_k['depressed'].map(map_depressed_int)


# Sleep issues
map_sleep = {
    '1': 'Yes', #'<=4h',
    '2': 'Yes', #'5h',
    '3': 'No', #'6h',
    '4': 'No', #'7h',
    '5': 'No', #'8h',
    '6': 'No', #'9h',
    '7': 'Yes' #'>=10h'
}
df_y21['sleep_issues_normalized'] = df_y21['sleep_issues'].map(map_sleep)
df_y23['sleep_issues_normalized'] = df_y23['sleep_issues'].map(map_sleep)

scale_sleep_to_bool = {
    1: 'No',
    2: 'No',
    3: 'No',
    4: 'Yes',
    5: 'Yes'
}
df_k['sleep_issues_normalized'] = df_k['sleep_issues'].map(scale_sleep_to_bool)

```

```python
# Seleccionamos aquellas variables que se pueden unir

df_k_normalized = df_k[['year', 'age_normalized', 'gender', 'occupation', 'stop_daily_activities_normalized', 'social_use', 'depressed_normalized', 'sleep_issues_normalized']]
df_k_normalized.columns = ['year', 'age', 'gender', 'occupation', 'stop_daily_activities', 'social_use', 'depressed', 'sleep_issues']

df_y21_normalized = df_y21[['year', 'age_normalized', 'gender_normalized', 'occupation_normalized', 'stop_daily_activities_normalized', 'social_use_normalized', 'depressed_normalized', 'sleep_issues_normalized']]
df_y21_normalized.columns = ['year', 'age', 'gender', 'occupation', 'stop_daily_activities', 'social_use', 'depressed', 'sleep_issues']

df_y23_normalized = df_y23[['year', 'age_normalized', 'gender_normalized', 'occupation_normalized', 'stop_daily_activities_normalized', 'social_use_normalized', 'depressed_normalized', 'sleep_issues_normalized']]
df_y23_normalized.columns = ['year', 'age', 'gender', 'occupation', 'stop_daily_activities', 'social_use', 'depressed', 'sleep_issues']


# Añadimos la fuente original al dataset como una columna, para poder tener un mejor tracking
df_k_normalized['source'] = 'Kaggle'
df_y21_normalized['source'] = 'YRBS21'
df_y23_normalized['source'] = 'YRBS23'
```

```python
print(df_k_normalized.head())
print(df_y21_normalized.head())
print(df_y23_normalized.head())
```

```python
# Concatenar datasets
df_combined = pd.concat([df_k_normalized, df_y21_normalized, df_y23_normalized], ignore_index=True)

df_combined.head()
```

### Análisis exploratorio

```python
# EDA variables categóricas
def graficos_eda_categoricos(cat):

    #Calculamos el número de filas que necesitamos
    from math import ceil
    filas = ceil(cat.shape[1] / 2)

    #Definimos el gráfico
    f, ax = plt.subplots(nrows = filas, ncols = 2, figsize = (16, filas * 6))

    #Aplanamos para iterar por el gráfico como si fuera de 1 dimensión en lugar de 2
    ax = ax.flat

    #Creamos el bucle que va añadiendo gráficos
    for cada, variable in enumerate(cat):
        cat[variable].value_counts().plot.barh(ax = ax[cada])
        ax[cada].set_title(variable, fontsize = 12, fontweight = "bold")
        ax[cada].tick_params(labelsize = 12)


graficos_eda_categoricos(df_combined_cleaned.select_dtypes('O'))
```

Dado que para la variable de género no disponemos de suficientes casos para los géneros no-binarios (únicamente 7), se deciden eliminar del dataset final

```python
print(df_combined_cleaned[~df_combined_cleaned['gender'].isin(['Male', 'Female'])].shape[0])

df_combined_cleaned = df_combined_cleaned[df_combined_cleaned['gender'].isin(['Male', 'Female'])]
```

```python
# EDA variables numéricas
def estadisticos_cont(num):
    #Calculamos describe
    estadisticos = num.describe().T
    #Añadimos la mediana
    estadisticos['median'] = num.median()
    #Reordenamos para que la mediana esté al lado de la media
    estadisticos = estadisticos.iloc[:,[0,1,8,2,3,4,5,6,7]]
    #Lo devolvemos
    return(estadisticos)

estadisticos_cont(df_combined_cleaned.select_dtypes('number'))
```

No vemos nada raro en la variable de edad, el rango es plausible con un mínimo de 12 años y un máximo de 91; además de tener una media de 16 años.  
Dado que los cuestionarios de Youth Risk Behaviour representan la mayoría del dataset, es comprensible.

### Generación de Insights

#### Insight1: ¿Hay alguna correlación entre un uso excesivo de las redes sociales y factores psicológicos?

```python
# Creamos una variable booleana para depressed
df_combined_cleaned['depressed_bool'] = df_combined_cleaned['depressed'].map({'Always': 1, 'Most of the time': 1, 'Sometimes': 1, 'Rarely': 0, 'Never': 0})
```

```python
# Análisis de depresión por social use
temp = df_combined_cleaned.groupby('social_use').depressed_bool.mean().sort_values(ascending = False) * 100
ax = temp.plot.bar()

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.title('Porcentaje de personas con depresión por Uso de redes sociales')
plt.show()
```

```python
# Creamos una variable booleana para stop daily activities
df_combined_cleaned['stop_daily_activities_bool'] = df_combined_cleaned['stop_daily_activities'].map({'Yes': 1, 'No': 0})

# Análisis de stop daily activities por social use
temp = df_combined_cleaned.groupby('social_use').stop_daily_activities_bool.mean().sort_values(ascending = False) * 100
ax = temp.plot.bar()

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.title('Porcentaje de personas que paran sus actividades diarias por Uso de redes sociales')
plt.show()
```

Podemos observar algo interesante, y es que alrededor del 70% de personas que consumen entre 4 y 5 horas diarias de redes sociales son más propensas a sufrir de trastornos por depresión.

#### Insight 2: ¿Afectan las redes sociales al sueño?

```python
# Creamos una variable booleana para stop daily activities
df_combined_cleaned['sleep_issues_bool'] = df_combined_cleaned['sleep_issues'].map({'Yes': 1, 'No': 0})

# Análisis de stop daily activities por social use
temp = df_combined_cleaned.groupby('social_use').sleep_issues_bool.mean().sort_values(ascending = False) * 100
ax = temp.plot.bar()

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.title('Porcentaje de personas con problemas de sueño por Uso de redes sociales')
plt.show()
```

De nuevo, el grupo de personas que utilizan entre 4 y 5h las redes sociales, más de la mitad sufren algún tipo de problema para dormir.

#### Insight 3: ¿Cuál es la diferencia, si la hay, entre géneros y el uso excesivo de las redes sociales?

```python
df_combined_cleaned['social_abuse_bool'] = df_combined_cleaned['social_use'].map({
    'Less than an Hour': 0,
    'Between 1 and 2 hours': 0,
    'Between 2 and 3 hours': 0,
    'Between 3 and 4 hours': 1,
    'Between 3 and 4 hours': 1,
    'More than 5 hours': 1
})

# Análisis de uso excesivo de redes sociales por género
temp = df_combined_cleaned.groupby('gender').social_abuse_bool.mean().sort_values(ascending = False) * 100
ax = temp.plot.bar()

ax.set_xticklabels(ax.get_xticklabels(), rotation=45, ha='right')
plt.title('Porcentaje de personas con uso excesivo de RRSS por Género')
plt.show()
```

Podemos observar que prácticamente el porcentaje de mujeres que tienen algún problema de uso excesivo de redes sociales, frente al de los hombres, es nulo.
