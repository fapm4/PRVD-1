## 1. Importación y preprocesamiento de datos

```python
import pyodbc
import pandas as pd

# Ruta al archivo .mdb
mdb_file = "datos/XXH2023_YRBS_Data.mdb"

# Crear la conexión con pyodbc
conn = pyodbc.connect(
    r"Driver={Microsoft Access Driver (*.mdb, *.accdb)};"
    f"DBQ={mdb_file};"
)

# Consulta SQL para leer datos de una tabla específica
query_q = "SELECT * FROM XXHq"  # set de preguntas Q


# Ejecutar la consulta y cargar los datos en un DataFrame
df_q = pd.read_sql(query_q, conn)


# Cerrar la conexión
conn.close()

# Mostrar las primeras filas del DataFrame
df_q.head()

```

```python
# Cogemos las variables que vamos a estudiar
df = df_q[['q1', 'q2', 'q25', 'q80']].copy()
df.columns = ['age', 'genre', 'electronic_bullying', 'social_use']
df
```

```python
# Diccionarios de mapeo para cada columna
map_edad = {
    '1': '<=12 years',  
    '2': '13 years',
    '3': '14 years',
    '4': '15 years',
    '5': '16 years',
    '6': '17 years',
    '7': '>=18 years'
}

map_genero = {'1': 'Female', '2': 'Male'}
map_acoso = {'1': 'Yes', '2': 'No'}
map_uso = {
    '1': 'I do not use social media',
    '2': 'A few times a month',
    '3': 'About once a week',
    '4': 'A few times a week',
    '5': 'About once a day',
    '6': 'Several times a day',
    '7': 'About once an hour',
    '8': 'More than once an hour'
}

# Crear un diccionario general para mapear cada columna
mapeos = {
    'age': map_edad,
    'genre': map_genero,
    'electronic_bullying': map_acoso,
    'social_use': map_uso
}


# Aplicar los mapeos a las columnas del DataFrame
for columna, mapeo in mapeos.items():
    df[columna] = df[columna].replace(mapeo)

# Mostrar el DataFrame con los valores categóricos
df.head()
```

```python
import matplotlib.pyplot as plt

# Lista de columnas categóricas
columnas_categoricas = ['edad', 'género', 'acoso por redes sociales', 'uso de redes sociales']

# Mostrar los datos
# Crear gráficos de barras por cada columna categórica
for columna in df.columns:
    plt.figure(figsize=(8, 6))
    df[columna].value_counts().plot(kind='bar', color='skyblue', edgecolor='black')
    plt.title(f'Frecuencia de {columna}', fontsize=14)
    plt.xlabel(columna, fontsize=12)
    plt.ylabel('Frecuencia', fontsize=12)
    plt.xticks(rotation=45, fontsize=10)
    plt.grid(axis='y', linestyle='--', alpha=0.7)
    plt.tight_layout()
    plt.show()
```

## 2. Relación entre las variables

### 2.1. Bullying electrónico

```python
df_bullied = df[df['electronic_bullying'] == 'Yes']
df_bullied
```

```python

# Contar la frecuencia de cada edad
edad_counts = df_bullied['age'].value_counts().sort_index()

# Graficar
plt.bar(edad_counts.index, edad_counts.values, color='skyblue')
plt.title('Distribución de edades (estudiantes que han sufrido bullying)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.xticks(edad_counts.index)  # Asegura que las etiquetas de las barras estén correctamente posicionadas
plt.show()
```

```python
# Contar la frecuencia de cada género
genero_counts = df_bullied['genre'].value_counts()

# Graficar
plt.bar(genero_counts.index, genero_counts.values, color='lightcoral')
plt.title('Distribución por género (estudiantes que han sufrido bullying)')
plt.xlabel('Género')
plt.ylabel('Frecuencia')
plt.show()

```

```python
# Contar la frecuencia de cada categoría de uso de redes sociales
uso_redes_counts = df_bullied['social_use'].value_counts()

# Graficar
plt.bar(uso_redes_counts.index, uso_redes_counts.values, color='mediumseagreen')
plt.title('Uso de redes sociales (estudiantes que han sufrido bullying)')
plt.xlabel('Categoría de uso')
plt.ylabel('Frecuencia')
plt.xticks(rotation=45)  # Rota las etiquetas del eje X si son largas
plt.show()
```

```python
# Crear una tabla cruzada
edad_genero = df_bullied.pivot_table(index='age', columns='genre', aggfunc='size', fill_value=0)

# Graficar
edad_genero.plot(kind='bar', stacked=False, color=['lightblue', 'pink'], figsize=(10, 6))
plt.title('Distribución de Edad por Género (estudiantes que han sufrido bullying)')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
plt.legend(title='Género')
plt.show()
```
